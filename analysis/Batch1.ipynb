```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data25.csv')
df
```

```python
# voltage
plt.figure(figsize=(12,9))
for i in range(len(df)) : 
      for j in range(i, len(df)) :
            if(df['wireGauge'][i] == df['wireGauge'][j] and
               df['turns'][i] == df['turns'][j] and 
               df['radius'][i] == df['radius'][j] and 
               df['wireGauge_car'][i] == df['wireGauge_car'][j] and
               df['turns_car'][i] == df['turns_car'][j] and 
               df['radius_car'][i] == df['radius_car'][j] and 
               df['height'][i] == df['height'][j] and 
               df['spacing'][i] == df['spacing'][j] and 
               df['velocity'][i] == df['velocity'][j]):
                        v = [float(df['V'][i]),float(df['V'][j])]
                        charge = [float(df['totCharge'][i]),float(df['totCharge'][j])]
                        plt.plot(v, charge,color='blue')
plt.title('Output Charge vs. Voltage')
plt.xlabel("Voltage [V]")
plt.ylabel("Output Charge [C]")
```
